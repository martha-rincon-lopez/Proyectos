```python
df['work_year'] = df.work_year.astype(str)
df.info()
```

Para definir categorias, como jerarquias

```python
from pandas.api.types import CategoricalDtype

categorias = CategoricalDtype(categories = ['EN', 'MI', 'SE', 'EX'], ordered = True) #Definimos la categoría
df['experience_level'] = df.experience_level.astype(categorias)
df.info()
```

Para saber cuánto hay de cada categorias

```python
df.value_counts('employment_type')
```

Porcentaje dependiendo el nivel de experiencia

```python
df.value_counts('experience_level', normalize = True)
```

Agregamos una columna, axis es para comparar por columna

```python
df['same_country'] = df.apply(lambda x: x.employee_residence == x.company_location, axis =1 )
df
```
